# **基本迴圈練習**

```
sum=0
n=int(input("請輸入正整數"))
for i in range(1,n+1):
  sum+=i
print("1到%d的整數和為:%d"%(n,sum))
```

## **九九乘法**

```
for i in range(1,10):
  for j in range(1,10):
    product=i*j
    print('%d * %d =%2d '%(j,i,product),end='  ')
    
  print('\n')
```

# **質數判斷**

```
n=int(input('請輸入大於1的整數:'))
count=1
if n<=1:
  print('error')
elif n==2:
  print('2是質數')
else:
  for i in range(2,n):
    if(n%i==0):
      print('不是質數')
      break
  if(n%i!=0) :  
    print('是質數')
```

# **成績輸入**

```
total = person = score = 0
while(score != -1):
    person += 1
    total += score
    score = input("請輸入第%d位學生的成績:" % person)
    score = int(score)
average = total / (person - 1)
print("本班總成績:%d分,平均成績:%5.2f分" % (total, average))

  
```

# **串列練習(數字輸入)**

```
score = []
total = inscore = 0
while inscore !=-1 :
  inscore = int(input('請輸入學生的成績:'))
  score.append(inscore)
print('共有%d位學生'%(len(score)-1))
for i in range(0,len(score)-1):
  total+=score[i]
average = total / (len(score)-1)
print('本班總分為: %d分 ,平均成績%5.2f分 '%(total,average))
```

# **串列練習(字典)**

```
dict1 = {'林小明':85,'曾山水':93,'鄭美麗':67}
dict1['黃明品']=71
dict1['陳莉莉']=98
listkey = list(dict1.keys());
listvalue = list(dict1.values())
for i in range(len(listkey)):
  print('%s的成績為%d分'%(listkey[i],listvalue[i])) 
```

```
dict1 = {"林小明":85, "曾山水":93, "鄭美麗":67}
dict1["黃明品"] = 71
dict1["陳莉莉"] = 98
listitem = dict1.items()
for name, score in listitem:
    print("%s的成績為%d分" % (name, score))
```
